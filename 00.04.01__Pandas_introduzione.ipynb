````python
import matplotlib.pyplot as plt
import pandas as pd
import wget
````

Legiamo i dati sulla pandemia covid 19 dal repository <a href="https://github.com/pcm-dpc/COVID-19">GitHub</a> del Dipartimento di Protezione Civile. È anche possibile importare da un file in formato xlsx.

````python
wget.download('https://raw.githubusercontent.com/pcm-dpc/COVID-19/master/dati-regioni/dpc-covid19-ita-regioni.csv', out='data')
````

````python
df = pd.read_csv('data/dpc-covid19-ita-regioni.csv')
````

Grafici con mathplotlib

````python
plt.figure()
df_campania.plot(x='data',y='terapia_intensiva',figsize=(12,10))
````

Manipolazione delle date

````python
df['data']
````

````python
df['data_dt'] = pd.to_datetime(df['data'])
````

````python
df[['data', 'data_dt']]
````

````python
day_1 = df['data_dt'].min()
````

````python
day_1
````

````python
df['giorni'] = (df['data_dt'] - day_1).dt.days
````

````python
df_campania = df.loc[df1['denominazione_regione']=="Campania"]
````

````python
df_regione = df.loc[(df['denominazione_regione']=='Piemonte') & (df['giorni']>200)] 
plt.figure()
fig, ax = plt.subplots()
plt.yscale('log')
df_regione.plot(x='giorni',y='terapia_intensiva', ax=ax, figsize=(12,10))
````
